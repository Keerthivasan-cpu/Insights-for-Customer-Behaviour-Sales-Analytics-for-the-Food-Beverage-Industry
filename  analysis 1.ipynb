**Step 1: Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 2: Load the Datasets**

```python
transactions = pd.read_csv("fb_retail_transactions.csv")
customers = pd.read_csv("fb_customer_master.csv")
products = pd.read_csv("fb_product_master.csv")
```

**Step 3: Check the Data**

```python
print(transactions.head())

print(customers.head())

print(products.head())
```

```python
print(transactions.info())
print(customers.info())
print(products.info())
```

```python
print(transactions.shape)
print(customers.shape)
print(products.shape)
```

**Step 4: Merge the Datasets**

```python
df = pd.merge(transactions, customers, on="customer_id", how="left")
```

```python
df = pd.merge(df, products, on="product_id", how="left")
```

**Step 5: Verify**

```python
print(df.head())
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
print(df.drop_duplicates())
```

**Part 2 – NumPy Analysis**

```python
import numpy as np
```

**Step 2: Select the Revenue Column**

```python
sales = df["net_revenue"]
print(sales.head(10))
```

**1. Average Revenue**

```python
median = np.median(sales)

print("Median Revenue:", median)
```

**2.Maximum Revenue**

```python
max=np.max(sales)
print("Maximum Revenue:",max)
```

**3.Minimum Sales**

```python
min=np.min(sales)
print("Minimum Revenue:",min)
```

**4.Standard Deviation**

```python
std=np.std(sales)
print("Standard Deviation:",std)
```

**5.Sum of Revenue**

```python
tot=np.sum(sales)
print("Total Revenue:",tot)
```

**7.Revenue Percentiles**

```python
print("25th Percentile:", np.percentile(sales,25))
print("50th Percentile:", np.percentile(sales,50))
print("75th Percentile:", np.percentile(sales,75))
```

**8.Revenue Range**

```python
range_value = np.max(sales) - np.min(sales)

print("Revenue Range:", range_value)
```

**9.Correlation Between Revenue & Sales**

```python
correlation = np.corrcoef(df["net_revenue"],df["profit"])
print(correlation)
```

**10.Complete NumPy Summary**


```python
print("Total Revenue :", np.sum(sales))
print("Average Revenue :", np.mean(sales))
print("Median Revenue :", np.median(sales))
print("Maximum Revenue :", np.max(sales))
print("Minimum Revenue :", np.min(sales))
print("Standard Deviation :", np.std(sales))
print("25th Percentile :", np.percentile(sales,25))
print("75th Percentile :", np.percentile(sales,75))
```
