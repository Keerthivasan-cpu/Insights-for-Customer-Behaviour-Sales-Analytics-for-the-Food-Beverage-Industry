```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/fb_customer_master.csv")

df.head()
```

**1. Bar Chart (Matplotlib)**

How many customers belong to each loyalty tier?

```python
tier = df["loyalty_tier"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(tier.index, tier.values)

plt.title("Customers by Loyalty Tier")
plt.xlabel("Loyalty Tier")
plt.ylabel("Number of Customers")

plt.show()
```

**2. Pie Chart (Matplotlib)**


Gender Distribution

```python
gender = df["gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    gender.values,
    labels=gender.index,
    autopct="%1.1f%%"
)

plt.title("Gender Distribution")

plt.show()
```

3. Horizontal Bar Chart

```python
city = df["city"].value_counts()

plt.figure(figsize=(8,5))
plt.barh(city.index, city.values)

plt.title("Customers by City")
plt.xlabel("Customers")
plt.ylabel("City")

plt.show()
```

**4. Line Chart**

Suppose you want to knowcustomer registrations over time.

```python
df["join_date"] = pd.to_datetime(df["join_date"])

join = df.groupby(df["join_date"].dt.month).size()

plt.figure(figsize=(8,5))

plt.plot(join.index, join.values, marker="o")

plt.title("Monthly Customer Registrations")
plt.xlabel("Month")
plt.ylabel("Customers")

plt.show()
```

**5. Histogram**

Age group frequency

```python
plt.figure(figsize=(6,4))

df["age_group"].value_counts().plot(kind="bar")

plt.title("Age Group Distribution")

plt.show()
```

age group vs city

```python
pd.crosstab(df["age_group"], df["city"]).plot(kind="bar", figsize=(10,5))

plt.title("Age Group vs City")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=55)
plt.legend(title="city")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pandas: Analyze data
loyality = df["loyalty_tier"].value_counts()

# Matplotlib: Plot graph
plt.bar(loyality.index, loyality.values,color="orange")

plt.title("Loyalty Tier Distribution")
plt.xlabel("Loyalty Tier")
plt.ylabel("Count")
plt.legend(loyality.values)
plt.show()
```

Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/fb_customer_master.csv")

table = pd.crosstab(df["gender"], df["loyalty_tier"])

plt.figure(figsize=(6,4))

sns.heatmap(table, annot=True, cmap="Blues", fmt="d")

plt.title("Gender vs Loyalty Tier")
plt.xlabel("Loyalty Tier")
plt.ylabel("Gender")

plt.show()
```
